```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("8-fraud_detection.csv")
```

```python
df.columns
```

```python
#imbalanced dataset
```

```python
X=df.drop("is_fraud",axis=1)
y=df["is_fraud"]
```

```python
sns.scatterplot(x=X["transaction_amount"],y=X["transaction_risk_score"],hue=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"]
c_values = [100,10,1.0,0.1,0.01]
solver = ["newton-cg","lbfgs","liblinear","sag","saga","newton-cholesky"]
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
class_weight
```

```python
params = dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
```

```python
cv=StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model,param_grid=params,scoring="accuracy",cv=cv)
```

```python
import warnings
warnings.filterwarnings("ignore")
#uyarı mesjalrını göstermez
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("score: ",accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print("confusion matrix: \n ",confusion_matrix(y_pred,y_test))
```

```python
grid.best_params_
```

ROC (Receiver Operating Characteristic) Eğrisi
İkili sınıflandırma (binary classification) problemlerinde model performansını görselleştirmek için kullanılır.
X ekseni: False Positive Rate (FPR) → Yanlış alarm oranı
Y ekseni: True Positive Rate (TPR) = Recall → Doğru yakalama oranı
ROC eğrisi, modelin farklı eşik değerlerinde (threshold) nasıl davrandığını gösterir.
Eğer eşik çok düşükse → model çoğu şeyi "1" der → TPR artar ama FPR da artar.
Eğer eşik çok yüksekse → model çoğu şeyi "0" der → TPR düşer ama FPR de düşer.
Yani ROC eğrisi aslında modelin duyarlılık (sensitivity) ile yanlış alarm (false positive) arasındaki dengeyi gösterir.

2. AUC (Area Under Curve)

ROC eğrisinin altında kalan alanın ölçüsüdür.
Modelin genel ayrıştırma gücünü özetleyen tek bir sayı verir.
0.5 → Model tamamen rastgele tahmin yapıyor (yani yazı-tura atmak gibi).
1.0 → Model mükemmel şekilde sınıfları ayırıyor.
0.7 - 0.8 → İyi
0.8 - 0.9 → Çok iyi
0.9+ → Harika

✅ Özet:

ROC → Doğru pozitif oranı ile yanlış pozitif oranını görselleştiren eğri.

AUC → O eğrinin altında kalan alan, modelin genel başarısını özetleyen değer.

```python
model_prob = grid.predict_proba(X_test)
```

```python
model_prob
```

```python
model_prob =model_prob[:,1]
#probabilities for the positive(fraud)class(pozitif sınıf için olan olasılıkları verir)
```

```python
model_prob
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
model_auc = roc_auc_score(y_test,model_prob)
```

```python
model_auc
```

```python
#model false positive rate
#model true positive rate
model_fpr,model_tpr,thresholds=roc_curve(y_test,model_prob)
```

```python
plt.plot(model_fpr,model_tpr,marker = ".",label="Logistic")
plt.xlabel("False Positive Rate")
plt.ylabel("Ture Positive Rate")
plt.show()
#çizginin altındaki alan ne kadar büyükse modelimizin performansı o kadar iydir
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(model_fpr,model_tpr,marker=".",label="Logistic")
for fpr,tpr,thresh in zip (model_fpr,model_tpr,thresholds):
    ax.annotate(f"{np.round(thresh,2)}",(fpr,tpr),ha="center")  
plt.plot(model_fpr,model_tpr,marker = ".",label="Logistic")
plt.xlabel("False Positive Rate")
plt.ylabel("Ture Positive Rate")
plt.show()
```
