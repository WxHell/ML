```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("8-fraud_detection.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df["is_fraud"].unique()
```

```python
df["is_fraud"].value_counts()
```

```python
#imbalanced dataset
```

```python
X=df.drop("is_fraud",axis=1)
y=df["is_fraud"]
```

```python
sns.scatterplot(x=X["transaction_amount"],y=X["transaction_risk_score"],hue=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"]
c_values = [100,10,1.0,0.1,0.01]
solver = ["newton-cg","lbfgs","liblinear","sag","saga","newton-cholesky"]
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
class_weight
```

```python
params = dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
```

```python
cv=StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model,param_grid=params,scoring="accuracy",cv=cv)
```

```python
import warnings
warnings.filterwarnings("ignore")
#uyar mesjalrn gstermez
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("score: ",accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print("confusion matrix: \n ",confusion_matrix(y_pred,y_test))
```

```python
grid.best_params_
```

ROC (Receiver Operating Characteristic) Erisi
kili snflandrma (binary classification) problemlerinde model performansn grselletirmek iin kullanlr.
X ekseni: False Positive Rate (FPR)  Yanl alarm oran
Y ekseni: True Positive Rate (TPR) = Recall  Doru yakalama oran
ROC erisi, modelin farkl eik deerlerinde (threshold) nasl davrandn gsterir.
Eer eik ok dkse  model ou eyi "1" der  TPR artar ama FPR da artar.
Eer eik ok yksekse  model ou eyi "0" der  TPR der ama FPR de der.
Yani ROC erisi aslnda modelin duyarllk (sensitivity) ile yanl alarm (false positive) arasndaki dengeyi gsterir.

2. AUC (Area Under Curve)

ROC erisinin altnda kalan alann lsdr.
Modelin genel ayrtrma gcn zetleyen tek bir say verir.
0.5  Model tamamen rastgele tahmin yapyor (yani yaz-tura atmak gibi).
1.0  Model mkemmel ekilde snflar ayryor.
0.7 - 0.8  yi
0.8 - 0.9  ok iyi
0.9+  Harika

 zet:

ROC  Doru pozitif oran ile yanl pozitif orann grselletiren eri.

AUC  O erinin altnda kalan alan, modelin genel baarsn zetleyen deer.

```python
model_prob = grid.predict_proba(X_test)
```

```python
model_prob
```

```python
model_prob =model_prob[:,1]
#probabilities for the positive(fraud)class(pozitif snf iin olan olaslklar verir)
```

```python
model_prob
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
model_auc = roc_auc_score(y_test,model_prob)
```

```python
model_auc
```

```python
#model false positive rate
#model true positive rate
model_fpr,model_tpr,thresholds=roc_curve(y_test,model_prob)
```

```python
plt.plot(model_fpr,model_tpr,marker = ".",label="Logistic")
plt.xlabel("False Positive Rate")
plt.ylabel("Ture Positive Rate")
plt.show()
#izginin altndaki alan ne kadar bykse modelimizin performans o kadar iydir
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(model_fpr,model_tpr,marker=".",label="Logistic")
for fpr,tpr,thresh in zip (model_fpr,model_tpr,thresholds):
    ax.annotate(f"{np.round(thresh,2)}",(fpr,tpr),ha="center")  
plt.plot(model_fpr,model_tpr,marker = ".",label="Logistic")
plt.xlabel("False Positive Rate")
plt.ylabel("Ture Positive Rate")
plt.show()
```
